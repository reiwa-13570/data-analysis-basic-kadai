```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, tight_layout = True)

# 折れ線グラフ
x0 = [0, 0.5, 1, 2]
y1 = [1, 2, -4, 2]
axes[0,0].plot(x0,y1)

# sin関数
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

fig, ax = plt.subplots()

axes[0,1].plot(x, y, linewidth=1.0)

axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

# ヒストグラム
import numpy as np
import matplotlib.pyplot as plt

x2 = np.random .normal(15, 5, 1860)

axes[0, 2].hist(x2)

# 散布図
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 60)
y = 6 + np.random.normal(0, 2.5, len(x))

axes[1, 0].scatter(x, y)

plt.show()
```
